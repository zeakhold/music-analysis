```python
n = note.Note("C4")
print(n.name)
print(n.octave)
print(n.pitch)
print(n.pitch.frequency)
print(n.pitch.pitchClass)
print(n.pitch.pitchClassString)

n.duration.quarterLength = 1
print(n.duration)

n.addLyric(n.nameWithOctave)
n.addLyric(n.pitch.pitchClassString)
n.show()
```

```python
bflat = note.Note("C4")
print(bflat.name)
print(bflat.octave)
if bflat.pitch.accidental is not None:
    print(bflat.pitch.accidental.name)
    print(bflat.pitch.accidental)
    print(bflat.pitch.accidental.alter)
    print(bflat.pitch.accidental.displayLocation)
# acc = bflat.pitch.accidental
bflat.show()
```

```python
d = bflat.transpose("M3")
print(bflat, d)
print(bflat.octave, d.octave)
print(bflat.pitch.frequency, d.pitch.frequency)
d.show()
```

```python
n = bflat.transpose("dd2")
print(bflat, n)
print(bflat.octave, n.octave)
print(bflat.pitch.frequency, n.pitch.frequency)
n.show()
```

```python
s = stream.Stream()
ts = meter.TimeSignature('3/4')
s.append(ts)
s.append(note.Note('C4', type='quarter'))
s.append(note.Note('D4', type='half'))
s.append(note.Note('E4', type='quarter'))
s.append(note.Note('F4', type='quarter'))
s.append(note.Note('G4', type='half'))

print(ts.beatCount)

for eachNote in s.notes:
    eachNote.addLyric(eachNote.beat)
#     print(note.beat)

s.show()
# print(s.notes[0].beat, s.notes[1].beat, s.notes[2].beat, s.notes[3].beat, s.notes[4].beat)
```
