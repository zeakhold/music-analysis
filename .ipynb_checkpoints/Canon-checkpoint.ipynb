```python
from music21 import *
import os
import numpy as np
import scipy as sp
from scipy.signal import fftconvolve
import matplotlib.mlab
import matplotlib.pyplot as plt
import pylab
import wave
import struct
# import music21
```

```python
readFile = 'data/wav/canon_2.wav'

#################################
# 读取音频
#################################
f = wave.open(readFile,'rb')
params = f.getparams()
nchannels, sampwidth, framerate, nframes = params[:4]
print(params)
# print(nchannels, sampwidth, framerate, nframes)
# nframes = audioSearch.audioChunkLength
# nframes = framerate * 1

#################################
# WaveData
#################################
strData = f.readframes(nframes) #读取音频，字符串格式
f.close()
print('strData', len(strData))
waveData = np.fromstring(strData, dtype=np.int16) #将字符串转化为int
print('waveData:', type(waveData), waveData.shape, np.max(waveData), np.min(waveData), np.mean(waveData))
print('waveData:', waveData[0:20])

waveData.shape = -1,2
waveData = waveData.T
print('waveData:', type(waveData), waveData.shape, np.max(waveData), np.min(waveData), np.mean(waveData))
print('waveData left:', waveData[0][0:10])
print('waveData right:', waveData[1][0:10])
# waveData = waveData*1.0/(max(abs(waveData))) #wave幅值归一化


#################################
# 1. AutoCorrelation
#################################
# correlationLeft = audioSearch.autocorrelationFunction(waveData[0], framerate)
# correlationRight = audioSearch.autocorrelationFunction(waveData[1], framerate)
# print('\nauto correlation 1:', correlationLeft, correlationRight)
# print(audioSearch.normalizeInputFrequency(correlationLeft), audioSearch.normalizeInputFrequency(correlationRight))

freqFromAQList = []
for i in range(int(nframes / audioSearch.audioChunkLength)):
    start = i * audioSearch.audioChunkLength
    end = start + audioSearch.audioChunkLength
    freqLeft = audioSearch.autocorrelationFunction(waveData[0][start:end], framerate)
    freqRight = audioSearch.autocorrelationFunction(waveData[1][start:end], framerate)
    freqFromAQList.append(freqLeft)
    if i < 10:
        print('\nauto correlation:', freqLeft, freqRight)
        print(audioSearch.normalizeInputFrequency(freqLeft), audioSearch.normalizeInputFrequency(freqRight))


#################################
# 2. FreqFromAudioFile
#################################
# freqFromAQList = audioSearch.getFrequenciesFromAudioFile(waveFilename=readFile)
chrome = scale.ChromaticScale('C4')
detectedPitchesFreq = audioSearch.detectPitchFrequencies(freqFromAQList, useScale=chrome)
print('\nfreqFromAQList:', len(freqFromAQList))
print(freqFromAQList[0:10])
print('\ndetectedPitchesFreq:', len(detectedPitchesFreq))
print(np.unique(detectedPitchesFreq))

(detectedPitches, listplot) = audioSearch.pitchFrequenciesToObjects(detectedPitchesFreq, useScale=chrome)
print('\ndetectedPitches:', len(detectedPitches))
print(detectedPitches[0:10])
print('\nlistplot:', len(listplot))
print(listplot[0:10])

(notesList, durationList) = audioSearch.joinConsecutiveIdenticalPitches(detectedPitches)
print('\nnotesList:', len(notesList))
print(notesList)
print('\ndurationList:', len(durationList))
print(durationList)
print('\nquarterLengthEstimation:', audioSearch.quarterLengthEstimation(durationList))

for i in range(len(notesList)):
    duration = audioSearch.quantizeDuration(durationList[i]/24)
    notesList[i].duration.quarterLength = duration
#     print(notesList[i].pitch.nameWithOctave, audioSearch.quantizeDuration(durationList[i]/24))

print(durationList)
s,lengthPart = audioSearch.notesAndDurationsToStream(notesList, durationList)
print('lengthPart:', lengthPart)
for n in s.parts[0].getElementsByClass('Note'):
    n.addLyric(n.pitch.nameWithOctave)
    n.addLyric(n.duration.quarterLength)
s.show()


#################################
# 3. FFT
#################################



#################################
# plot the wave
#################################
time = np.arange(0, nframes)*(1.0 / framerate)
plt.figure(figsize=(20,8))
plt.subplot(211)
plt.ylabel("Amplitude")
plt.title("Wave Left")
plt.grid('on')
plt.plot(time, waveData[0], color = 'green')
plt.subplot(212)
plt.xlabel("Time(s)")
plt.ylabel("Amplitude")
plt.title("Wave Right")
plt.grid('on')
plt.plot(time, waveData[1])
plt.show()

#################################
# FFT
#################################
# 采样点数，修改采样点数和起始位置进行不同位置和长度的音频波形分析
N = nframes
start = 0 #开始采样位置
df = framerate/(N-1) # 分辨率
freq = [df*n for n in range(0,N)] #N个元素
wave_data2 = waveData[0]#[start:start+N]
c = np.fft.fft(wave_data2)*2/N

#常规显示采样频率一半的频谱
d = int(len(c)/2)

# #仅显示频率在4000以下的频谱
while freq[d]>4000:
    d-=10
pylab.plot(freq[:d-1],abs(c[:d-1]),'r')
pylab.show() 


#################################
# specgram
#################################
# # framerate就是44100，文件初读取的值。然后本程序最关键的一步！specgram！实在太简单了
# Fs = framerate  
# pylab.specgram(waveData[0], NFFT=1024, Fs=Fs, noverlap=900)  
# pylab.show()  
```
